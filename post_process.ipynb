## Participant Editing Style, Need for Alternative Words, and Current Writing Aids

```python
# Editing Style
with open("./jsons/editing_style.json", "r", encoding="utf-8") as file:
    editing_data = json.load(file)

freq = collections.Counter(editing_data.values())

# Create bar plot
plt.figure(figsize=(6, 4))
ax = sns.barplot(x=list(freq.keys()), y=list(freq.values()), palette="pastel")

# Add labels on top of bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.xlabel("Stage of Writing")
plt.ylabel("Frequency")
plt.title("Participant Editing Styles")
plt.show()
```

## SYNthia Words

```python
# Set the directory containing the JSON files
directory = "./jsons/official"
novice_participants = []
expert_participants = []

num_words_prompted_novice = []
num_words_prompted_expert = []
num_words_prompted_all = []

percent_words_accepted_novice = [] # have this as percent: num_words_accepted_novice/num_words_prompted_novice
percent_words_accepted_expert = [] # have this as percent: num_words_accepted_expert/num_words_prompted_expert
percent_words_accepted_all = []

# Loop through each file in the directory
for filename in os.listdir(directory):
    if filename.endswith(".json"):  # Ensure it's a JSON file
        file_path = os.path.join(directory, filename)
        
        with open(file_path, "r", encoding="utf-8") as file:
            data = json.load(file)
            # print(f"{filename}: {len(data)}")
            num_words_prompted_all.append(len(data))
            words_accepted = sum(bool(entry["word_chosen"]) for entry in data)
            percent_words_accepted_all.append(round(float(words_accepted/len(data)), 2))

            if 'novice' in filename:
                novice_participants.append(filename.split('_')[0])
                num_words_prompted_novice.append(len(data))

                # add num words accepted
                words_accepted = sum(bool(entry["word_chosen"]) for entry in data)
                percent_words_accepted_novice.append(round(float(words_accepted/len(data)), 2))
            else:
                expert_participants.append(filename.split('_')[0])
                num_words_prompted_expert.append(len(data))

                # add num words accepted
                words_accepted = sum(bool(entry["word_chosen"]) for entry in data)
                percent_words_accepted_expert.append(round(float(words_accepted/len(data)), 2))
```

```python
n_data = {'Group': ['Novice'] * len(num_words_prompted_novice) + ['Expert'] * len(num_words_prompted_expert),
          'Words Prompted': num_words_prompted_novice + num_words_prompted_expert}
n_df = pd.DataFrame(n_data)

p_data = {'Group': ['Novice'] * len(percent_words_accepted_novice) + ['Expert'] * len(percent_words_accepted_expert),
          'Percent Accepted': percent_words_accepted_novice + percent_words_accepted_expert}
p_df = pd.DataFrame(p_data)

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# First plot: Number of Words Prompted
sns.boxplot(x='Group', y='Words Prompted', data=n_df, ax=axes[0], showfliers=False, width=0.5, boxprops={'zorder': 1}, palette="pastel")
sns.swarmplot(x='Group', y='Words Prompted', data=n_df, ax=axes[0], color='black', size=5, zorder=2)
axes[0].set_title("Number of Words Prompted")

# Second plot: Percent of Words Accepted
sns.boxplot(x='Group', y='Percent Accepted', data=p_df, ax=axes[1], showfliers=False, width=0.5, boxprops={'zorder': 1}, palette="pastel")
sns.swarmplot(x='Group', y='Percent Accepted', data=p_df, ax=axes[1], color='black', size=5, zorder=2)
axes[1].set_title("Percent of Words Accepted")

plt.tight_layout()
plt.show()
```

```python
# Independent samples T-test with 1 factor (percent words accepted) and 2 levels (novice, experts)
t_stat, p_value = stats.ttest_ind(percent_words_accepted_novice, percent_words_accepted_expert, equal_var=False)
print(f"T-statistic: {t_stat:.3f}, P-value: {p_value:.3f}")
```

```python
# Set the directory containing the JSON files
novice_ai_participants = ['p1', 'p2', 'p3', 'p4', 'p6', 'p9', 'p11', 'p14', 'p15', 'p19']
expert_ai_participants = ['p5', 'p7', 'p8', 'p10', 'p12', 'p13', 'p16', 'p17', 'p18']
# novice_ai_participants = ['p2', 'p3', 'p4', 'p6', 'p7', 'p9', 'p11', 'p12', 'p15', 'p17', 'p18', 'p19']
# expert_ai_participants = ['p1', 'p5', 'p8', 'p10', 'p13', 'p14', 'p16']
directory = "./jsons/official"

num_words_prompted_novice_ai = []
num_words_prompted_expert_ai = []

percent_words_accepted_novice_ai = [] # have this as percent: num_words_accepted_novice/num_words_prompted_novice
percent_words_accepted_expert_ai = [] # have this as percent: num_words_accepted_expert/num_words_prompted_expert

# Loop through each file in the directory
for filename in os.listdir(directory):
    if filename.endswith(".json"):  # Ensure it's a JSON file
        file_path = os.path.join(directory, filename)
        
        with open(file_path, "r", encoding="utf-8") as file:
            data = json.load(file)
            # print(f"{filename}: {len(data)}")
            participant = filename.split('_')[0]

            if participant in novice_ai_participants:
                num_words_prompted_novice_ai.append(len(data))

                # add num words accepted
                words_accepted = sum(bool(entry["word_chosen"]) for entry in data)
                percent_words_accepted_novice_ai.append(round(float(words_accepted/len(data)), 2))
            else:
                num_words_prompted_expert_ai.append(len(data))

                # add num words accepted
                words_accepted = sum(bool(entry["word_chosen"]) for entry in data)
                percent_words_accepted_expert_ai.append(round(float(words_accepted/len(data)), 2))
```

```python
n_data = {'Group': ['Novice'] * len(num_words_prompted_novice_ai) + ['Expert'] * len(num_words_prompted_expert_ai),
          'Words Prompted': num_words_prompted_novice_ai + num_words_prompted_expert_ai}
n_df = pd.DataFrame(n_data)

p_data = {'Group': ['Novice'] * len(percent_words_accepted_novice_ai) + ['Expert'] * len(percent_words_accepted_expert_ai),
          'Percent Accepted': percent_words_accepted_novice_ai + percent_words_accepted_expert_ai}
p_df = pd.DataFrame(p_data)

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# First plot: Number of Words Prompted
sns.boxplot(x='Group', y='Words Prompted', data=n_df, ax=axes[0], showfliers=False, width=0.5, boxprops={'zorder': 1}, palette="pastel")
sns.swarmplot(x='Group', y='Words Prompted', data=n_df, ax=axes[0], color='black', size=5, zorder=2)
axes[0].set_title("Number of Words Prompted (Agg by AI skill)")

# Second plot: Percent of Words Accepted
sns.boxplot(x='Group', y='Percent Accepted', data=p_df, ax=axes[1], showfliers=False, width=0.5, boxprops={'zorder': 1}, palette="pastel")
sns.swarmplot(x='Group', y='Percent Accepted', data=p_df, ax=axes[1], color='black', size=5, zorder=2)
axes[1].set_title("Percent of Words Accepted (Agg by AI skill)")

plt.tight_layout()
plt.show()
```

```python
# Independent samples T-test with 1 factor (number words prompted) and 2 levels (novice, experts)
t_stat, p_value = stats.ttest_ind(num_words_prompted_novice_ai, num_words_prompted_expert_ai, equal_var=False)
print(f"T-statistic: {t_stat:.3f}, P-value: {p_value:.3f}")
```

```python
# Independent samples T-test with 1 factor (percent words accepted) and 2 levels (novice, experts)
t_stat, p_value = stats.ttest_ind(percent_words_accepted_novice_ai, percent_words_accepted_expert_ai, equal_var=False)
print(f"T-statistic: {t_stat:.3f}, P-value: {p_value:.3f}")
```

## Time Spent on SYNthia vs. Writing

```python
novice_synthia_times = [309, 62, 58, 200, 90, 310, 255, 122, 147, 262, 100, 240]
novice_writing_times = [591, 838, 842, 700, 810, 590, 645, 778, 753, 638, 800, 660]
expert_synthia_times = [65, 110, 86, 94, 513, 227, 315]
expert_writing_times = [835, 790, 814, 806, 387, 673, 585]
```

```python
time_data = {
    'Group': ['Novice - Writing'] * len(novice_writing_times) + ['Expert - Writing'] * len(expert_writing_times) +
             ['Novice - SYNthia'] * len(novice_synthia_times) + ['Expert - SYNthia'] * len(expert_synthia_times),
    'Time (s)': novice_writing_times + expert_writing_times + novice_synthia_times + expert_synthia_times
}
time_df = pd.DataFrame(time_data)
```

```python
# Box plot
fig, ax = plt.subplots(figsize=(12, 6))
sns.boxplot(x='Group', y='Time (s)', data=time_df, ax=ax, showfliers=False, width=0.5, boxprops={'zorder': 1}, palette="pastel")
# Swarm plot
sns.swarmplot(x='Group', y='Time (s)', data=time_df, ax=ax, color='black', size=5, zorder=2)

ax.set_title("Writing and SYNthia Time Comparison")
# plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# Violin plot for Writing Times
sns.violinplot(x='Group', y='Time (s)', data=time_df[time_df['Group'].str.contains("Writing")], ax=axes[0], inner="quartile", palette="pastel")
axes[0].set_title("Writing Time Comparison")

# Violin plot for SYNthia Times
sns.violinplot(x='Group', y='Time (s)', data=time_df[time_df['Group'].str.contains("SYNthia")], ax=axes[1], inner="quartile", palette="pastel")
axes[1].set_title("SYNthia Time Comparison")

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

# Strip plot with jitter
sns.stripplot(x='Group', y='Time (s)', data=time_df, jitter=True, size=8, alpha=0.7, palette="pastel")

ax.set_title("Writing and SYNthia Time Comparison")
plt.tight_layout()
plt.show()
```
